```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
#nltk.download('vader_lexicon')
try:
    nltk.data.find('vader_lexicon')
except LookupError:
    nltk.download('vader_lexicon')
    print("vader_lexicon descargado")

analizador = SentimentIntensityAnalyzer()

def obtener_sentimiento(texto):

    puntuacion = analizador.polarity_scores(texto)
    score = puntuacion['compound']

    if score >= 0.05:
       # print('positivo')
        return 'positivo'
    elif score <= -0.05:
        return 'negativo'
        #print('negativo')
    else:
        #print('neutro')
        return 'neutro'

frase1 = "I love the service, the delivery was very fast, the app works wonderfully, it's excellent."
frase2 = "The service was very bad, the delivery was inefficient, I don't like it."
frase3 = "Delivery starts at 10 am."

print(obtener_sentimiento(frase1))
print(obtener_sentimiento(frase2))
print(obtener_sentimiento(frase3))

#0.05 o mayor positivo
#-0.05 valor negativo
#0 neutro
```
